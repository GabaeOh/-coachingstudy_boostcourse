## 4. 분석과 시각화를 위한 tidy data 만들기
- melt는 행에 있는 데이터를 열로 옮기는 것
    - d_vars로 id값으로 사용할 열을 지정하고, value_vars에는 value으로 들어가는 열을 지정
    - id_vars에 들어가지 않은 열들은 value_vars

```python
df_raw.melt(id_vars=["국가(대륙)별", "상품군별", "판매유형별"])
```

```python
df = df_raw.melt(id_vars=["국가(대륙)별", "상품군별", "판매유형별"], var_name="기간", value_name="백만원")
df.shape
```

```python
df[:3]
## 값에 "합계"가 많으므로 전처리 해야함
```

```python
df.info()
## 연도와 분기를 분리하고, 숫자로 변경해야 함
```

## 5. 데이터 전처리 

### 연도 분리 

```python
df['기간']
```

```python
# split()을 사용하면 공백으로 데이터를 분리
"2019 4/4 p)".split()
```

```python
"2019 4/4 p)".split()[0] #첫번째 인덱스를 가져오기
```

```python
type("2019 4/4 p)".split()[0])
```

```python
int("2019 4/4 p)".split()[0]) # integer 형으로 변경
```

```python
df["기간"].map?
```

```python
# map 안에 lambda 함수를 넣습니다. x.split()한 데이터의 첫번째 인덱스 값을 integer 형으로 변경
df["연도"] = df["기간"].str.extract(r'(\d{4})').astype(int)
df
```

### 분기 분리

```python
"2019 4/4 p)".split()[1].split("/") # 공백과 /을 기준으로 데이터를 split하여 4/4에 있는 4와 4를 분리
```

```python
"2019 4/4 p)".split()[1].split("/")[0] # 4/4에 있는 앞의 4를 가져오기
```

```python
df["분기"] = list(map(lambda x : int(x.split(".")[1].split()[0].split("/")[0]), df["기간"]))
df.head()
```

### 금액을 수치 데이터로 표현하기 위해 데이터 타입 변경

```python
## '백만원' 컬럼의 '-'을 NaN으로 대체
df["백만원"].replace("-", np.nan, inplace=True)
```

```python
# 숫자는 소숫점으로 변경
df["백만원"] = df["백만원"].replace("-", np.nan).astype(float)
df["백만원"]
```

### 필요 없는 데이터 제거

```python
df.info()
```

```python
# 값이 "합계"가 아닌 데이터만 집계하여 df을 다시 업데이트
df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
df
```

```python
df.isnull().sum() #금액이 없는 데이터도 있다는 것을 참고
```

## 6. K-Beauty 시각화

### 6.1 전체 상품군 판매액

```python
# NaN으로 표시된 결측치 데이터가 있으면 시각화가 잘 안나타날 수 있으므로 NaN으로 나타난 행의 데이터를 제거
df_total = df[df["판매유형별"] == "계"].copy()
df_total.head()
```

```python
# 연도에 따른 판매액을 lineplot
sns.lineplot(data=df_total, x="연도", y="백만원")
```

```python
sns.lineplot(data=df_total, x="연도", y="백만원", hue="상품군별")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # legend 값을 밖에 표시하는 소스코드
```

```python
sns.relplot(data=df_total, x="연도", y="백만원", hue="상품군별", kind="line")
```

```python
sns.relplot(data=df_total, x="연도", y="백만원", hue="상품군별", kind="line", col="상품군별", col_wrap=4)
## 화장품이 다른 상품군들에 비해 해외직접판매액이 꾸준히 증가하고 있다는 것을 알 수 있습니다.
```

```python
df_sub = df_total[~df_total["상품군별"].isin(["화장품"])].copy() # '~' 상품군별이 "화장품"인 데이터만 빼고 df_sub 변수
df_sub
```

```python
sns.relplot(data=df_sub, x="연도", y="백만원", hue="상품군별", col="상품군별", col_wrap=4, kind="line")
## 의류, 패션 상품도 많이 판매함을 알 수 있습니다. 가전 전자 통신기기와 음반 비디오 악기 쪽도 성장세가 있음을 알 수 있습니다.
```

```python
df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션관련 상품"])].copy()
```

```python
sns.relplot(data=df_sub, x="연도", y="백만원", hue="상품군별", col="상품군별", col_wrap=4, kind="line") # 의류, 패션 상품 빼고
```

### 6.2 화장품의 온라인쇼핑 해외직접판매액

```python
df_cosmetic = df_total[df_total["상품군별"] == "화장품"].copy()
df_cosmetic
```

```python
df_cosmetic["상품군별"].unique()
```

```python
sns.lineplot(data=df_cosmetic, x="연도", y="백만원")
## 2019년까지 꾸준히 성장하고 있습니다.
```

```python
plt.figure(figsize=(15, 4))
sns.lineplot(data=df_cosmetic, x="연도", y="백만원", hue="분기")
## 1분기부터 4분기까지 계속 성장하고 있습니다.
```

### 기간별 화장품 판매액 데이터 시각화

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30) # 글씨를 겹쳐지지 않게 하기 위해 xticks()를 사용하여 글자를 회전
sns.lineplot(data=df_cosmetic, x="기간", y="백만원")
```

```python
df_cosmetic.head()
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간",  y="백만원", hue="국가(대륙)별")
## 중국에서의 가장 판매액이 높습니다.
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] != "중국"], 
             x="기간", y="백만원", hue="국가(대륙)별")
# 중국을 빼고 시각화 그래프를 그리면 아세안에서 최근 높은 판매량을 기록했음을 알 수 있습니다.
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[df["판매유형별"] != "계"].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별")
## "계" 데이터를 빼고 그래프를 그리니 온라인 면세점이 성장하고 있다는 사실을 알 수 있습니다.
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"] != "면세점")].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별", ci=None)
## 온라인 면세점도 빼고 시각화, 면세점 이외에서도 증가 추세를 볼 수 있습니다.
```

### 6.3 의류 및 패션 관련 상품 온라인쇼핑 해외직접판매액 시각화

```python
df["상품군별"].value_counts()
```

```python
df_fashion = df[df["상품군별"] == "의류 및 패션 관련상품"].copy()
df_fashion.head()
```

```python
df_fashion = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
# 기간별 금액 데이터를 시각화하고, 국가(대륙별)을 색상으로 표시
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="백만원", hue="국가(대륙)별")
## 꾸준히 증가하고 있습니다. 중국, 일본, 미국, 아세안 순으로 성장하고 있습니다.
```

```python
df_fashion2 = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] != "계")].copy() #"계"를 제거 

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x="기간", y="백만원", hue="판매유형별", ci=None)
# 면세점 이외의 데이터는 꾸준히 성장했고, 면세점 데이터도 성장하고 있습니다. 두 부문 모두 계속 성장 중입니다.
```

### 6.4 데이터 집계하기 
- pivot_table
- heatmap
    - 분포를 색상으로 편하게 볼 수 있습니다.

```python
df_fashion
```

```python
df_fashion.pivot_table(index="국가(대륙)별", columns="연도", values="백만원") #기본값이 mean
```

```python
df_fashion.pivot_table(index="국가(대륙)별", columns="연도", values="백만원", aggfunc="sum")
```

```python
df_fashion["판매유형별"].value_counts()
# 계만 있기 때문에 이 데이터를 가공 없이 그대로 sum을 사용하면 연도별 합계를 알 수 있습니다.
```

```python
result = df_fashion.pivot_table(index="국가(대륙)별", columns="연도", values="백만원", aggfunc="sum")
result
```

```python
sns.heatmap(result)
# vmax는 시각화를 할 때 나타내는 최대값입니다. 
# vmax와 vmin
```

```python
sns.heatmap(result, cmap = "Blues_r") # cmap 옵션으로 색상을 변경
```

```python
sns.heatmap(result, cmap = "Blues_r", annot=True) # annot 옵션으로 수치를 표시
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(result, cmap="Blues_r", annot=True, fmt=".0f")# fmt float형의 숫자를 나타낼 수 있습니다.
## 중국에서 구매액이 가장 많습니다. 미국과 일본에서도 구매액이 늘어나고 있습니다.
```
