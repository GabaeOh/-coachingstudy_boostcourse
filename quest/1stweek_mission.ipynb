## 1. 설문조사 문항의 응답내역을 분석
- 문항별 응답내역이 'question'에 담겨 있을 때, 조건문과 반복문을 사용하여 아래와 같은 결과가 출력되도록 코드를 작성해보세요.

- 문항별 응답내용에는 하나의 응답만 할 수 있는 single choice 문제와 여러 응답을 선택할 수 있는 multiple choice 문제가 있습니다.

## 2. 연도별, 국가별 평균 기대수명 출력
- 2011년 부터의 연도별, 국가별 평균 기대수명을 구해주세요. (2011년도 포함되게 구합니다.)
- [필수 조건] groupby 나 pivot_table을 활용합니다. groupby 로 구할 때 unstack()이라는 기능을 사용하여 아래와 같이 컬럼에 인덱스 값을 올려서 표기할 수 있습니다.
- 만약 행에는 '연도'가, 열에 '국가'가 들어있고 수치 데이터의 결과값이 아래와 같다면, 출력형태는 조금 달라도 괜찮습니다.

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
#2011년부터의 데이터만 확보
df = df[df['Year'] >= 2011]
df[:3]
```

```python
# 'Year'와 'Country'를 기준으로 'Life_Expectancy'의 평균값을 출력 
df_result = df.groupby(['Year', 'Country'])['Life_Expectancy'].agg(['mean']).unstack()
df_result
```
